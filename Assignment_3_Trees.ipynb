```python
# 2.  Find height of a given tree

class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key
 
 
# A function to do inorder tree traversal and find height
def printInorder(root):
    if root is None:
        return 0
    else:
        # Find the height of left and right subtrees
        left_height = printInorder(root.left)
        right_height = printInorder(root.right)

        # Return the maximum height plus one (to account for the current node)
        return 1 + max(left_height, right_height)
 
# Driver code
if __name__ == "__main__":
    root = Node(1)
    root.left = Node(2)
    root.right = Node(3)
    root.left.left = Node(4)
    root.left.right = Node(5)
    root.right.left = Node(6)
    root.right.right = Node(7)
    
    # Find the height of the tree
    tree_height = printInorder(root)
    
    # Print the height
    print("Height of the binary tree is", tree_height)
```

```python
# 5.  Implement BFS (Breath First Search) and DFS (Depth First Search)

class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

def bfs(root):
    queue = [root]
    while queue:
        node = queue.pop(0)
        print(node.val)
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)

def dfs_preorder(node):
    if node is None:
        return
    print(node.val)
    dfs_preorder(node.left)
    dfs_preorder(node.right)

def dfs_inorder(node):
    if node is None:
        return
    dfs_inorder(node.left)
    print(node.val)
    dfs_inorder(node.right)

def dfs_postorder(node):
    if node is None:
        return
    dfs_postorder(node.left)
    dfs_postorder(node.right)
    print(node.val)

if __name__ == "__main__":
    root = Node(1)
    root.left = Node(2)
    root.right = Node(3)
    root.left.left = Node(4)
    root.left.right = Node(5)
    root.right.left = Node(6)
    root.right.right = Node(7)

print("BFS:")
bfs(root)  # Output: 1 2 3 4 5 6 7

print("DFS (preorder):")
dfs_preorder(root)  # Output: 1 2 4 5 3 6 7

print("DFS (inorder):")
dfs_inorder(root)  # Output: 4 2 5 1 3 6 7

print("DFS (postorder):")
dfs_postorder(root)  # Output: 4 5 2 7 6 3 1
```

```python
# 6.  Find sum of all left leaves in a given Binary Tree

class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

def printInorder(root):

    if root:
      printInorder(root.left)
      print(root.val)
      printInorder(root.right)

def sum_left_leaves(root):
    if root is None:
        return 0

    # Check if the left child is a leaf node and add its value
    if root.left is not None and root.left.left is None and root.left.right is None:
        return root.left.val + sum_left_leaves(root.right)

    # Otherwise, recursively traverse both subtrees
    return sum_left_leaves(root.left) + sum_left_leaves(root.right)


if __name__ == "__main__":
    root = Node(1)
    root.left = Node(2)
    root.right = Node(3)
    root.left.left = Node(4)
    root.left.right = Node(5)
    root.right.left = Node(6)
    root.right.right = Node(7)
    
 
    # Function call
    print ("\nInorder traversal of binary tree is")
    printInorder(root)

su = sum_left_leaves(root)
print("Sum of Left Leaves is :",su)
```

```python
# 7.  Find sum of all nodes of the given perfect binary tree

class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

def printInorder(root):

    if root:
      printInorder(root.left)
      print(root.val)
      printInorder(root.right)

def sum_nodes(root):
    if root is None:
        return 0

    # Calculate the sum of the left and right subtrees recursively
    left_sum = sum_nodes(root.left)
    right_sum = sum_nodes(root.right)

    # Add the value of the current node to the sum of both subtrees
    return root.val + left_sum + right_sum


if __name__ == "__main__":
    root = Node(1)
    root.left = Node(2)
    root.right = Node(3)
    root.left.left = Node(4)
    root.left.right = Node(5)
    root.right.left = Node(6)
    root.right.right = Node(7)
    
 
    # Function call
    print ("\nInorder traversal of binary tree is")
    printInorder(root)

sum_node = sum_nodes(root)
print("Sum of all the Nodes is: ",sum_node)
```

```python
# 9.  Find maximum level sum in Binary Tree

class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key
 
 
def printInorder(root):
 
    if root:
        printInorder(root.left)
        print(root.val)
        printInorder(root.right)

def max_level_sum(root):
    if not root:
        return 0

    max_sum = float('-inf')
    current_sum = 0
    level = 0
    queue = [root]

    while queue:
        level += 1
        current_sum = sum(node.val for node in queue)
        max_sum = max(max_sum, current_sum)
        next_level = []

        for node in queue:
            if node.left:
                next_level.append(node.left)
            if node.right:
                next_level.append(node.right)

        queue = next_level

    return max_sum

# Driver code
if __name__ == "__main__":
    root = Node(1)
    root.left = Node(2)
    root.right = Node(3)
    root.left.left = Node(4)
    root.left.right = Node(5)
    root.right.left = Node(6)
    root.right.right = Node(7)
    
 
    # Function call
    print ("\nInorder traversal of binary tree is")
    printInorder(root)

max_sum = max_level_sum(root)
print(f"Maximum level sum: {max_sum}")  # Output: 11
```

```python
# 10. Print the nodes at odd levels of a tree


class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key
 
 
def printInorder(root):
 
    if root:
        printInorder(root.left)
        print(root.val)
        printInorder(root.right)

def print_odd_levels(root):
    if not root:
        return

    queue = [root]
    level = 1

    while queue:
        if level % 2 == 1:
            print([node.val for node in queue])

        next_level = []

        for node in queue:
            if node.left:
                next_level.append(node.left)
            if node.right:
                next_level.append(node.right)

        queue = next_level
        level += 1


# Driver code
if __name__ == "__main__":
    root = Node(1)
    root.left = Node(2)
    root.right = Node(3)
    root.left.left = Node(4)
    root.left.right = Node(5)
    root.right.left = Node(6)
    root.right.right = Node(7)
    
 
    # Function call
    print ("\nInorder traversal of binary tree is")
    printInorder(root)
    print ("\nnodes at odd levels of a tree")
    print_odd_levels(root)
```
